```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fernando.csv', encoding='cp1252', index_col=0)
df.head(10)
```

```python
duplicated = df.duplicated().sum()
print('Quantidade de linhas duplicadas: ', duplicated)
df.drop_duplicates(inplace=True)
```

**Substituindo valores nas colunas Supervisor e Orientador de 'Sim' para '1' e 'Não' para '0'**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Supervisor'] = le.fit_transform(df['Supervisor'])
df['Orientador'] = le.fit_transform(df['Orientador'])
df.head(10)
```

**Substituindo valores nas colunas 'Orientador' de '0' para '1'**

```python
data = df.replace({'Orientador': {0: 1}})
data.head(10)
```

**Criando pivot table para verificar a quantidade de alunos por modalidade**

```python
df.pivot_table(index=['Modalidade'], aggfunc='size')
```

**Criando pivot table para verificar a quantidade de alunos por agente de integração**

```python
df.pivot_table(index=['Agente Integrador'], aggfunc='size')
```

**Criando pivot table para comparar a quantidade de alunos orientados e supervisionados**
